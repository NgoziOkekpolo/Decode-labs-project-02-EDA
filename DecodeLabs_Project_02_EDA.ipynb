```python
#Bivariate Analysis - examines the relationship between two variables to identify patterns, trends and business insights
#Total Sales by Payment Method
totalsalesbyPM = df.groupby("PaymentMethod")["TotalPrice"].sum()
totalsalesbyPM
```

```python
#Total Sales by Referral Source
salesbyreferral = df.groupby("ReferralSource")["TotalPrice"].sum()
salesbyreferral
```

```python
#Average Order Value by Product
avgorderbyproduct = df.groupby("Product")["TotalPrice"].mean()
avgorderbyproduct
```

# Key Business Insights
Based on the exploratory data analysis(EDA), the following insights were identified:
1. Order Status - The distribution of order statuses is fairly balanced across all categories. However, Cancelled orders recorded the 
highest count, closely followed by Returned and Pending orders. This may suggest opportunities to investigate customer satisfaction,
delivery issues, or order processing efficiency.

2. Payment Method - Customers used all available payment methods almost equally, but Online Payment was the most frequently used option.
This indicates that customers are comfortable making digital payments on the platform.

3. Coupon Code Usage - Among customers who used discount codes, FREESHIP was the most popular coupon. This suggests that free shipping 
promotions may have a stronger influence on purchasing decisions than percentage discounts.

4. Referral Source - Instagram generated the highest number of customer visits, making it the most effective customer acquisition channel in 
this dataset. This suggests that marketing efforts on Instagram are producing positive results.
    
5. Total Price Distribution - The histogram shows that most purchases were concentrated in the lower to mid price range, while only a small 
number of orders had very high total prices. This indicates that high-value purchases occur less frequently.
    
6. Sales by Payment Method  - Analysis of total sales revealed that Credit Card payments generated the highest revenue. Although customers 
used different payment methods, credit card transactions contributed the most to overall sales.
    
7. Sales by Referral Source - Among all referral channels, Instagram generated the highest total sales. This further supports the observation
that Instagram is currently the strongest marketing channel for attracting valuable customers.

8. Average Order Value by Product - The Laptop category recorded the highest average order value. This is expected since laptops generally
have higher selling prices than the other products in the dataset.

9. Average Order Value by Referral Source - Customers referred through Facebook recorded the highest average order value, even though 
Instagram attracted more customers overall. This suggests that Facebook visitors tend to spend slightly more per order.

10. Correlation Analysis - The correlation matrix revealed:
• Unit Price has a strong positive relationship with Total Price (0.72), meaning higher-priced products generally result in higher order 
values.
• Quantity also has a moderate positive relationship with Total Price (0.62), indicating that purchasing more items increases the total 
amount spent.
• Quantity and Items in Cart show a moderate positive relationship (0.65), suggesting customers who add more items to their cart often 
purchase larger quantities.
• Overall, no unexpected relationships were observed among the numerical variables.
    
11. Outlier Detection - The box plots showed a few high-value outliers in the Total Price column. These represent unusually expensive 
orders rather than data entry errors, so they were retained for analysis.
No significant outliers were observed in the Unit Price and Quantity columns, indicating that these variables are fairly consistent across 
the dataset.

# Recommendations
Based on the findings from this exploratory data analysis(EDA), the following recommendations are suggested:
1. Invest More in Instagram Marketing - Since Instagram generated both the highest number of customers and the highest total sales, 
increasing marketing efforts on this platform could further improve customer acquisition and revenue.

2. Investigate High Cancellation Rates - Cancelled orders recorded the highest frequency among all order statuses. The business should 
investigate possible causes such as delayed deliveries, payment issues, or customer dissatisfaction to reduce order cancellations.

3. Continue Offering Free Shipping Promotions - The FREESHIP coupon was the most frequently used discount code, suggesting that customers 
respond well to free shipping incentives. The company should consider maintaining or expanding similar promotional campaigns.
    
4. Encourage Credit Card Payments - Credit card transactions generated the highest total sales. Offering incentives such as cashback, 
loyalty rewards, or exclusive discounts for credit card payments may help increase revenue.
    
5. Promote High-Value Products - Since laptops recorded the highest average order value, targeted marketing campaigns for premium products
could help increase overall sales revenue.

6. Monitor High-Value Transactions - Although only a few high-value outliers were observed in the Total Price column, these transactions 
should be monitored regularly to ensure they represent genuine customer purchases and not fraudulent activities.

7. Improve Customer Experience Across Referral Channels - While Instagram attracted the most customers, Facebook users had the highest 
average order value. The business should optimize marketing strategies for each platform based on customer spending behavior.

# Conclusion
This exploratory data analysis provided a better understanding of customer purchasing patterns, product performance, and sales trends
within the dataset. The analysis showed that Instagram was the leading referral source for both customer acquisition and total sales, 
while Facebook customers recorded the highest average order value. Credit card payments contributed the highest sales revenue, and laptops
had the highest average order value among all products.                                                                                        The correlation analysis revealed that Unit Price and Quantity have a positive influence on Total Price, which aligns with expected business
behaviour. Outlier detection also identified a few high-value transactions in the Total Price column, but these appear to be legitimate

Overall, the dataset was clean and suitable for analysis, allowing meaningful insights to be drawn. These findings can support business
